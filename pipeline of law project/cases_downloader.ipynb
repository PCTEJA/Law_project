```
with open('cases_list (2).txt') as f:
    dict1=f.read()
with open('cases_list (1).txt') as f1:
    dict2 = f1.read()
list1=dict2.split('\n')
list1.pop(71)
dict2={}
for i in list1:
    import re
    p=str('http://liiofindia.org/in/cases/cen/INSC/'+i+'\/\d+\.html')
    s=re.findall(p, dict1)
    dict2[int(i)]=s
len(dict2.values())
```

```
from collections import OrderedDict 
from bs4 import BeautifulSoup
for yr in range(1996, 1997):
  r=str(str(yr)+'_text.txt')
  for ii in list(OrderedDict.fromkeys(dict2[yr])) :
    browser2 = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    browser2.get(ii)
    page2 = BeautifulSoup(browser2.page_source,features="xml")
    import re
    for i in page2.find_all('p'):
        if(i.string!=None):
            if('LiiofIndia' in i.string or '............' in i.string):
                continue
            else:
                with open(r, 'a') as f:
                    f.write(i.string)
  print(yr, 'Iteration Completed!!! ')
```
